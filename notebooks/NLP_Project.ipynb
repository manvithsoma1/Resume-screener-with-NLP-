# Intelligent Resume Screening - Complete NLP Pipeline
This notebook serves as the academic foundation for the project. We will explore every NLP concept step-by-step.

```
import re
import string

def clean_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'<.*?>', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text

print(clean_text('Check my portfolio https://example.com!'))
```

## 1. Text Preprocessing & Normalization
**Definition:** The process of cleaning text data to remove noise.
**Why:** Models cannot understand unstructured noise like URLs or HTML tags.
**Implementation:**

```
# Import NLTK and tokenizers
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')

tokens = word_tokenize(clean_text('Hello world, this is a test.'))
print('Tokens:', tokens)
```

## 2. Feature Engineering (TF-IDF)
**Formula:** TF(t,d) = (Count of t in d) / (Total words in d). IDF(t) = log_e(Total documents / Documents with t in it).

```
from sklearn.feature_extraction.text import TfidfVectorizer
corpus = ['I love machine learning.', 'Machine learning is awesome.']
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names_out())
print(X.toarray())
```

## 3. Cosine Similarity
**Definition:** A measure of similarity between two non-zero vectors.
**Formula:** cos(theta) = A.B / ||A|| ||B||

```
from sklearn.metrics.pairwise import cosine_similarity
print(cosine_similarity(X))
```
